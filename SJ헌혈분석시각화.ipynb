```python
age30 = {'월':94253,'화':88096, '수':92083, '목':89176, '금':103573, '토':121069, '일':89417 }
age40 = {'월':85672,'화':79177, '수':82502, '목':79828, '금':91323, '토':106802, '일':84000 }
dayblood = pd.DataFrame( [age30, age40])
dayblood= dayblood.T
dayblood.columns=['30대','40대']
dayblood = dayblood.reset_index()
dayblood = dayblood.drop('index', axis=1)
dayblood
```

```python
# dayblood.index=['30대', '40대']
dayblood = dayblood.melt()
daybloodc=pd.DataFrame(['월','화','수','목', '금','토','일','월','화','수','목', '금','토','일'])
dayblood = pd.merge(dayblood, daybloodc, left_index=True, right_index=True)
dayblood['tip_cumsum'] = dayblood.groupby([0])['value'].cumsum(axis=0)
dayblood = dayblood.sort_values(by=['variable'], ascending=False)
dayblood
```

```python
# sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [30, 15]

ax = sns.barplot(x=0, y='tip_cumsum',hue='variable', data=dayblood, dodge=False, palette='mako') #, palette='tab2')
ax.set_xlabel("요일",fontsize=45)
ax.tick_params(labelsize=45)
ax.set_ylabel("총횟수 (연령대별 색깔다름)",fontsize=45)
ax.set_title('전국 요일별 3040 헌혈량', fontsize=45)
plt.legend(fontsize=30)
```
